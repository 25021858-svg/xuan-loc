```python
#W6D1
def txl(n):
    lst=[]
    for i in n:
        if i not in lst:
            lst.append(i)
        else:
            continue
    return lst
n=list(map(int,input().split()))
print(txl(n))
        
```

```python
#W6D2
def txl(n):
    sum=0
    lst=[]
    for i in n:
        sum=sum+i
        lst.append(sum)
    return lst
n=list(map(int,input().split()))
print(txl(n))
    
```
