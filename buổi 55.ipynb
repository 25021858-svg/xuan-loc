```python
#W5D3
from math import sqrt
def check(n):
    for i in range(2, int(sqrt(n)) + 1):
        if n % i == 0: 
            return False
    return n > 1
```

```python
#W5D4
def check(n):
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s == n
```

```python
#W5D6
mp = {}
def factorial(n):
    global mp
    if n == 0: return 1
    if n in mp: return mp[n]
    mp[n] = n * factorial(n - 1)
    return mp[n]
```

```python
#W5D7
def ktra_dau(c):
    if c == "+": return 1
    elif c== "-": return 2
    elif c == "*": return 3
    return 4


def calculator(s: str):
    st = []
    idx = 0
    for i in range(len(s)):
        if s[i] == " ":
            idx = i
            break 

    chuoi = s[:idx]
    st.append(float(chuoi))
    idx += 1
    dau = ktra_dau(s[idx])
    chuoi = ""
    for i in range(idx, len(s)):
        if s[i] == " ":
            if chuoi != "":
                if dau == 1:
                    st.append(float(chuoi))
                elif dau == 2:
                    st.append(-float(chuoi))
                elif dau == 3:
                    a = st[len(st) - 1]
                    st.pop()
                    b = float(chuoi)
                    a = a * b
                    st.append(a)
                else:
                    a = st[len(st) - 1]
                    st.pop()
                    b = float(chuoi)
                    a = a / b
                    st.append(a)
                chuoi = ""
            else: continue
        elif (s[i] >= '0' and s[i] <= '9') or s[i] == ".":
            chuoi += s[i]
        else: 
            if s[i + 1] >= "0" and s[i + 1] <= '9': chuoi += s[i]
            else: dau = ktra_dau(s[i])
    if dau == 1:
        st.append(float(chuoi))
    elif dau == 2:
        st.append(-float(chuoi))
    elif dau == 3:
        a = st[len(st) - 1]
        st.pop()
        b = float(chuoi)
        a = a * b
        st.append(a)
    else:
        a = st[len(st) - 1]
        st.pop()
        b = float(chuoi)
        a = a / b
        st.append(a)
    
    ans = 0
    for i in st:
        ans += i
    
    return float("{:.2f}".format(ans))

print(calculator("12 / 4"))
```

```python
#W5D8
def Hamming(x, y):
    x = x ^ y
    ans = 0
    while x > 0:
        ans += x % 2
        x = int(x / 2)
    return ans
```

```python
#W5D9
def tong(n):
    s = str(n)
    ans = 0
    for i in s:
        ans += int(i)
    return ans
```

```python
#W5D10
def dangcau(s1, s2):
    mp1 = {}
    mp2 = {}
    for i in range(len(s1)):
        if s1[i] in mp1:
            if mp1[s1[i]] != s2[i]: return "false"
        else: mp1[s1[i]] = s2[i]
        if s2[i] in mp2:
            if mp2[s2[i]] != s1[i]: return "false"
        else: mp2[s2[i]] = s1[i]
    return "true"

print(dangcau("egg", "add"))
```
