```python
#W10D1
a,b=map(int,input().split())
try:
    c=a/b
    print(c)
except ZeroDivisionError:
    print("không thể chia cho 0")
except ValueError:
    print("nhập số thực")
finally:
    print('hoàn thành chương trình')

    
```

```python
#W10D3
try:
    n=int(input())
    da_gap=set()
    while n!=1 and n not in da_gap:
        h=0
        da_gap.add(n)
        while n>0:
            k=n%10
            n=n//10
            h=h+k**2
        n=h
    if h==1:
        print("True")
    else: 
        print('False')
except ValueError:
    print("nhập số nguyên")

    


        
    
```
